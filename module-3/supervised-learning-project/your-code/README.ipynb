## Explicación del dataset

DataSet link: https://www.kaggle.com/dansbecker/nba-shot-logs

Hemos descargado el DataSet de tiros de la NBA en temporada regular 2014-2015. Cabe destacar ciertos aspectos.

- El DataSet contiene 128.069 valores. Estos contienen información en 21 columnas
- En este DataSet solamente están contabilizados los tiros efectuados en partidos de temporada regular (82 partidos por equipo según el formato NBA).
- Están solamente incluídos tiros de campo, es decir:

    - No se contabilizan los tiros libres.
    - No se contabilizan los tiros en formato 2+1 o 3+1. Explicación: Tiros de este formato se corresponden a tiros exitosos efectuados por un jugador inmediatamente después de recibir falta personal. Se contabiliza el tiro además del jugador tener un tiro libre extra.
 
### Detalle de cada columnas

1. **GAME_ID:** El identificador del partido jugado
2. **MATCHUP:** Partido donde se realizó ese tiro: equipos involucrados y fecha
- **LOCATION:** Si el partido es en casa 'H' o fuera de casa 'A'
- **W:** Resultado del partido, 'W' para victoria 'L' para derrota
- **FINAL_MARGIN:** Diferencia final de puntuación del partido
- **SHOT_NUMBER:** Número de tiro del jugador determinado en ese partido
- **PERIOD:** Periodo del partido (valores entre 1 y 4)
- **GAME_CLOCK:** Tiempo determinado en el que se realizó ese tiro (Máximo 12 ya que son periodos de 12 minutos en la NBA)
- **SHOT_CLOCK:** Tiempo de posesión donde se realizó ese tiro (Valores entre 0 y 24)
- **DRIBBLES:** Número de regates previos de un jugador antes de realizar el tiro
- **TOUCH_TIME:** Tiempo de posesión de balón del jugador previo al lanzamiento
- **SHOT_DIST:** Distancia a canasta en la realización del tiro
- **PTS_TYPE:** Valoración del tiro en puntos (Valores entre 2 y 3)
- **SHOT_RESULT:** Resultado del tiro 'made' para tiro exitoso, 'missed' si se ha fallado
- **CLOSEST_DEFENDER:** Nombre del defensor más cercano al tirador
- **CLOSEST_DEFENDER_PLAYER_ID:** ID del defensor más cercano
- **CLOSE_DEF_DIST:** Distancia a la que se encontraba el defensor en la realización del tiro
- **FGM:** Si el tiro ha entrado, valor '1' si este ha sido exitoso, '0' si por lo contrario el tiro ha sido fallado
- **PTS:** Puntuación asociada al tiro (Valores 0, 2 o 3)
- **player_name:** Nombre del jugador que ha realizado el tiro
- **player_id:** ID del jugador que ha realizado el tiro

## Objetivo del análisis

El objetivo de este análisis es demostrar si existe alguna probabilidad de predecir de si un tiro es exitoso o fallido dependiendo de ciertos parámetros (momento del partido, tiempo de posesión restante, driblings previos al tiro, distancia a canalasta, distancia del defensor más próximo...)

## Modelos realizados

1. Logistic regresion
2. K-Nearest Neighbors (k=3 y k=5)
3. DecisionTreeClassifier
4. Support Vector Classification
5. RandomForest
