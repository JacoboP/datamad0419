```python
import pandas as pd
import os
import requests
import quandl
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from dotenv import load_dotenv
from collections import Counter
import pandas_profiling
```

```python
DataFrame='tableau_project.csv'

def read(DataFrame):
    data = pd.read_csv(DataFrame)
    return data
```

```python
data = read(DataFrame)
data
```

```python
data = data.fillna('unknown')
data = data.set_index('Unnamed: 0')
```

```python
data.columns
data.info()
```

```python
data['vehicle_type_id'] = data['vehicle_type_id'].replace({'21620ea5749f2e0679a8c72c7fbafb9e': 'Vehicle 1', 
                                                           '077866c3fd1a75f51ca7f8eae166ae32': 'Vehicle 2'})

data['start_lat'] = data['start_lat'].astype(str)
data['start_lon'] = data['start_lon'].astype(str)
data['end_lat'] = data['end_lat'].astype(str)
data['end_lon'] = data['end_lon'].astype(str)
```

```python
data['end_state'].value_counts()
#quitar discount
```

```python
data = data.drop(data[data['end_state']=='not fount'].index)
data = data.drop(data[data['end_state']=='no show'].index)
data = data.drop(data[data['end_state']=='driver cancel'].index)

```

```python
data.head()
```

```python
data.columns
```

```python
colnames = ['journey_id', 'start_type', 'start_at', 'start_lat',
                'start_lon', 'end_at', 'end_lat', 'end_lon', 'end_state', 'price',
                'price_distance', 'price_duration', 'distance', 'duration', 'source',
                'rider_waiting_time', 'driver_waiting_time', 'price_supplements', 
                'rating']
data = pd.DataFrame(data, columns=colnames)

```
